```python
import pandas as pd
```

```python
pampers_PHVPs_search = pd.read_csv('g_ipc_at_issue.tsv',sep="\t", low_memory=False)
```

```python
pampers_PHVPs_search
```

```python
pampers_nappies_fabric = pampers_PHVPs_search[(pampers_PHVPs_search['section'] == 'D') & (pampers_PHVPs_search['ipc_class'] == '04') & (pampers_PHVPs_search['subclass'] == 'H')]
pampers_nappies_fabric
```

```python
pampers_nappies_fabric['patent_id'] = pd.to_numeric(pampers_nappies_fabric['patent_id'], error='coerce')
```

```python
pampers_nappies_fabric_remove_duplicate = pampers_nappies_fabric.drop_duplicates(subset='patent_id')
pampers_nappies_fabric_remove_duplicate
```

```python
PHVPs_prediction = pd.read_csv('predited_value.csv')
PHVPs_prediction 
```

```python
pampers_nappies_fabric_predicted = pampers_nappies_fabric.merge(PHVPs_prediction, how="left", on="patent_id")
pampers_nappies_fabric_predicted
```

```python
pampers_nappies_PHVPs = pampers_nappies_fabric_predicted[(pampers_nappies_fabric_predicted['predicted_ANN']==1)]
pampers_nappies_PHVPs
```

```python
pampers_nappies_PHVPs = pampers_nappies_PHVPs.drop_duplicates(subset='patent_id')
pampers_nappies_PHVPs
```
